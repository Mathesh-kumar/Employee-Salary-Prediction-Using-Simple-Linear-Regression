## **Simple Linear Regression**
#### Machine Learning model to predict the salary for a new employee based on his experience.
### **Steps**
  1. Importing needed librearies
  2. Data preprocessing
  3. Training simple linear regression
  4. Predicting test results
  5. Visualizing train results
  6. Visualizing test results

```
# 1. importing needed libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
# 2. data preprocessing
# 2.1. importing needed dataset

df = pd.read_csv('Salary_Data.csv')
df.head()
```

```
# 2.2. checking for missing values
df.isnull().sum()
```

```
# 2.3. splitting dependent & independent variable

X = df.iloc[:, :-1].values
y = df.iloc[:, 1].values
```

```
# 2.4. splitting dataset into train & test

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```
# 2.5. feature scaling
# if needed you can use this.

""" from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X_train = sc_x.fit_transform(X_train)
X_test = sc_x.transfrom(X_test)
sc_y = StandardScaler()
y_train = sc_y.fit_transform(y_train.reshape(-1,1)) """
```

```
# 3. training simple linear regression on train set.
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```
# 4. predicting test set results
y_pred = lr.predict(X_test)
```

```
# just compare real salary and predicted salary
pd.concat([pd.DataFrame(y_test, columns=['y_test']), pd.DataFrame(y_pred, columns=['y_pred'])], axis=1)
```

```
# 5. visualizing train set results

plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, lr.predict(X_train), color='blue')
plt.title('salary vs experience')
plt.xlabel('experience')
plt.ylabel('salary')
plt.show()
```

```
# 6. visualizing test set results

plt.scatter(X_test, y_test, color='red')
plt.plot(X_train, lr.predict(X_train), color='blue')
plt.title('salary vs experience')
plt.xlabel('experience')
plt.ylabel('salary')
plt.show()
```
